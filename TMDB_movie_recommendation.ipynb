```python
import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['Actors'][0]
```

```python
#from the above dictionary, we need only top three cast
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['Actors'] = movies['Actors'].apply(convert2)
movies.head()
```

```python
#here we need only that dictionary whose job is Director, and we need to 
#extract name from that dictionary

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies_dataset = movies[['id','title','overview','Actors','crew','keywords','genres']]
movies_dataset
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset.dropna(inplace=True)
```

```python
movies_dataset['overview']= movies_dataset['overview'].apply(lambda x: x.split())
```

```python
movies_dataset['genres'] = movies_dataset['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_dataset['keywords'] = movies_dataset['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_dataset['Actors'] = movies_dataset['Actors'].apply(lambda x:[i.replace(" ","") for i in x])
movies_dataset['crew'] = movies_dataset['crew'].apply(lambda x:[i.replace(" ","") for i in x])

movies_dataset.head()
```

```python
# Now let's create a new column called tags in which we will concatenate
# overview, genres, keywords, cast, crew

movies_dataset['tags'] = movies_dataset['overview']+movies_dataset['genres']+movies_dataset['keywords']+movies_dataset['Actors']+movies_dataset['crew']
```

```python
movies_dataset.head()
```

```python
final = movies_dataset[['id','title','tags']]
final.head()
```

```python
#convert the list of tags to string
final['tags']= final['tags'].apply(lambda x:" ".join(x))
final.head()
```

```python
final['tags'][0]
```

```python
#let's convert everything in lower case

final['tags'] = final['tags'].apply(lambda x: x.lower())
final.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
final['tags'] = final['tags'].apply(stem)
```

```python
final['tags'][0]
```

```python
#we will take 5000 words out of the string(formed after combining all the tags) and 
# we will remove the stop words(like - the, is, and, or etc.)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english' )

vectors = cv.fit_transform(final['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()  #these are the 5000 words we have taken
```

```python
# Eucledin distance is not a reliable measure in high dimension so we use cosine distance

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape #cosine distance of every movie with every movies
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]

#if we just sort the similarity vector then we will lose the index i.e. which 
# distance is related to which movie, but when we use the enumerate then we get
# index as well as distance. Now we sort in descending order(reverse=True) and
# on the basis of second index (key=lambda x:x[1])
# [1:6] gives the first value to 6th value i.e. 5 values total
```

```python
#the diagonal values of the above matrix will be 1 because the similarity of every
# movie with itself is 1

#now we will make a function which we take a movie from me and recommend top 5
#movies with maximum similarity to the given movie.

#for that we are making this function which will first of all find the index of the giveb
#movie in the dataset and from the index it will get the similarity vector and
# then sort that vector in descending order and return top 5

def recommend(movie):
    movie_index = final[final['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(final.iloc[i[0]].title)
```

```python
recommend('The Lord of the Rings: The Two Towers')
```
